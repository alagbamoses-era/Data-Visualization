## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

GroupByData = Clean_df.groupby(['Drug Regimen'])


TumorMean = GroupByData.mean()['Tumor Volume (mm3)']
TumorMean                         
TumorMedian = GroupByData.median()['Tumor Volume (mm3)']
TumorMedian
TumorVar = GroupByData.var()['Tumor Volume (mm3)']
TumorVar
Tumorstd = GroupByData.std()['Tumor Volume (mm3)']
Tumorstd
Tumorsem = GroupByData.sem()['Tumor Volume (mm3)']
Tumorsem

# creating a Dictionary


TumorDict = {'Mean': TumorMean, 'Median':TumorMedian, 'TumorVar': TumorVar, 'standard deviation': Tumorstd, 'SEM': Tumorsem}

# creating summary DataFrame

Tumor_df = pd.DataFrame(TumorDict)
Tumor_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
 
  
# Start by getting the last (greatest) timepoint for each mouse


# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
```

```python
CopyCleanData = Clean_df.copy()
CopyCleanData.head()
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin



#FinalFourReg = CopyCleanData.set_index("Drug Regimen").loc[["Capomulin","Ramicane", "Infubinol","Ceftamin"]].reset_index()
#FinalFourReg

FinalFourReg = CopyCleanData[CopyCleanData["Drug Regimen"].isin(["Capomulin", "Ramicane", "Infubinol", "Ceftamin"])]
FinalFourReg 

#Start by getting the last (greatest) timepoint for each mouse
GreatTimePoint = FinalFourReg.groupby('Mouse ID').Timepoint.max()
GreatTimePoint


# Merge this group df with the original dataframe to get the tumor volume at the last timepoint


GreatTimePoint_df = pd.merge(FinalFourReg, GreatTimePoint, on = ["Timepoint", "Mouse ID"])
GreatTimePoint_df.tail()
```

```python
final_tv = GreatTimePoint_df.groupby(['Drug Regimen', 'Mouse ID']).last()['Tumor Volume (mm3)']
#final_tv_df = final_tv.to_frame()
#final_tv_df

final = pd.DataFrame(final_tv)
final
```

```python
# Put treatments into a list for for loop (and later for plot labels)

TreatmentList = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]


# Create empty list to fill with tumor vol data (for plotting)
EmptyList = []


# Calculate the IQR and quantitatively determine if there are any potential outliers. 

#print(GreatTimePoint_df)
for element in TreatmentList:
    FinalTumorVol = GreatTimePoint_df.loc[GreatTimePoint_df['Drug Regimen']== element, 'Tumor Volume (mm3)' ]
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    #print(FinalTumorVol)
    
    # add subset 
    EmptyList.append(FinalTumorVol)
    
    # Determine outliers using upper and lower bounds

    # STEPS TO:
    # 1. Calculate the quantiles of 0.25, 0.50, 0.75
    # 2. Iqr = upper - lower
    # 3. lower_outlier = lower - (1.5*iqr)
    # 4. upper_outlier = upper + (1.5*iqr)
    #5. Filter the data bases on the upper and lower outliers

    Quantile = FinalTumorVol.quantile([0.25, 0.50, 0.75])
    Upper = Quantile[0.75]
    Lower = Quantile[0.25]

    iqr = Upper -Lower

    # to calculate lower outlier

    lower_outlier = Lower - (1.5*iqr)


    # To calculate Upper outlier
    upper_outlier = Upper + (1.5*iqr)

    # Filter the data bases on the upper and lower outliers

    Outliers = FinalTumorVol[(FinalTumorVol < lower_outlier) | (FinalTumorVol > upper_outlier)]
    print(element, Outliers)
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

#plt.boxplot(EmptyList)

fig1, ax1 = plt.subplots(figsize=(14,10))
ax1.set_title('Tumor Volume for each Drug Regimen',fontsize =16)
ax1.set_ylabel('Final Tumor Volume (mm3)',fontsize = 14)
ax1.set_xlabel('Drug Regimen',fontsize = 14)
plt.boxplot(EmptyList, widths = 0.4, patch_artist=True,vert=True, labels = TreatmentList)
plt.savefig("boxplot_final_tumor_vol.png", bbox_inches = "tight")

plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 

# To calculate correlation coefficient

VolTumourWeight = AvgCapTumor_df[['Weight (g)', 'Tumor Volume (mm3)']]
VolTumourWeight.corr()

# for mouse weight and average tumor volume for the Capomulin regim
```

```python
# Add the linear regression model
x_values = AvgCapTumor_df['Weight (g)']
y_values = AvgCapTumor_df['Tumor Volume (mm3)']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
print(line_eq)

fig1, ax1 = plt.subplots(figsize=(12, 8))
plt.scatter(x_values,y_values, s=90, color="green")
plt.plot(x_values,regress_values,"r-")
plt.title("Linear Regression Model", fontsize=20)
plt.xlabel("Mouse weight (g)", fontsize =14)
plt.ylabel("Average Tumor Volume (mm3)", fontsize=14)
plt.annotate(line_eq, xy=(16,45),  fontsize=16,color="blue")

plt.savefig("regression_avg_tumor_vol_and_weight.png", bbox_inches = "tight")
plt.show()
```
